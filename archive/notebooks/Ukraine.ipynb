```python
import pandas as pd
import yfinance as yf

# =========================
# Import European Defense ETF Prices
# =========================
start = pd.Timestamp("2025-01-01").date()
end = pd.Timestamp("2026-01-01").date()


def fetch_one(symbols, start, end):
    """Try a list of symbols; return the first non-empty Close series (tz-naive)."""
    for s in symbols:
        try:
            df = yf.download(
                s,
                start=start,
                end=end,
                auto_adjust=True,
                progress=False,
                threads=False,
            )
            if isinstance(df, pd.DataFrame) and "Close" in df and not df["Close"].dropna().empty:
                sr = df["Close"].dropna().copy()
                sr.index = pd.to_datetime(sr.index).tz_localize(None)
                sr.name = s
                return sr
        except Exception:
            pass
    return pd.Series(dtype=float)


# European defense-focused ETFs (with fallback listings)
nato_ucits = fetch_one(["NATO.L", "NATO.AS", "NATO.DE"], start, end)
nato_ucits.name = "NATO_UCITS"

vaneck_defense = fetch_one(["DFNS.DE", "DFNS.L"], start, end)
vaneck_defense.name = "VANECK_DEFENSE"

# Assemble prices
df_etf = pd.concat([nato_ucits, vaneck_defense], axis=1).dropna(how="all")

print("European Defense ETF Prices:")
display(df_etf.tail())
```
